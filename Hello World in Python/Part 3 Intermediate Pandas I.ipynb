```python
# check missing in each column
penguins.isna() # if null value display True
```

```python
# sum n/a value each column
penguins.isna().sum()
```

```python
# filter missing value in column sex
penguins[ penguins['sex'].isna() ]
```

```python
# drop na
clean_penguins = penguins.dropna()

clean_penguins.head(10)
```

```python
# fill missing value with mean imputation
top5_penguins = penguins.head(5)
```

```python
avg_value = top5_penguins['bill_length_mm'].mean()
print(avg_value)
```

```python
top5_penguins = top5_penguins['bill_length_mm'].fillna(value = avg_value)
top5_penguins
```

```python
# sort bill_length_mm low to high, high to low
#penguins.sort_values('bill_length_mm') # sort default is asc
#penguins.sort_values('bill_length_mm', ascending=False)
penguins.dropna().sort_values('bill_length_mm', ascending=False).head(10) # drop n/a before sort and display top 10 rows
```
