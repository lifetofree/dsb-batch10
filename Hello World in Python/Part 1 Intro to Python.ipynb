# OOP

```python
class Dog:
    name = "Phon"
    age = 5
    color = "Brown"
    breed = "French Bulldog"

    # function (Dog method)
    def sitting(self):
        print("I am sitting now!")

    def hungry(self, food_name):
        print(f"I am hungry, I need {food_name}")
```

```python
my_dog = Dog()
type(my_dog)
```

```python
my_dog.sitting()
```

```python
my_dog.hungry("Hotdog")
```
